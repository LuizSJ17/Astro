```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
M15 = pd.read_csv('/content/drive/MyDrive/+Estelar/Evolucionary tracks - 15Mo.csv', delimiter=';')
M20 = pd.read_csv('/content/drive/MyDrive/+Estelar/Evolucionary tracks - 20Mo.csv', delimiter=';')
M25 = pd.read_csv('/content/drive/MyDrive/+Estelar/Evolucionary tracks - 25Mo.csv', delimiter=';')
M30 = pd.read_csv('/content/drive/MyDrive/+Estelar/Evolucionary tracks - 30Mo.csv', delimiter=';')
M40 = pd.read_csv('/content/drive/MyDrive/+Estelar/Evolucionary tracks - 40Mo.csv', delimiter=';')
```

```python
plt.figure(figsize=(7,7))
plt.plot(M15.loc[200:,'log_Teff'],M15.loc[200:,'log_L'],'tab:red',label='15$M_{\odot}$')
plt.plot(M20.loc[200:,'log_Teff'],M20.loc[200:,'log_L'],'tab:orange',label='20$M_{\odot}$')
plt.plot(M25.loc[200:,'log_Teff'],M25.loc[200:,'log_L'],'tab:green',label='25$M_{\odot}$')
plt.plot(M30.loc[200:,'log_Teff'],M30.loc[200:,'log_L'],'tab:blue',label='30$M_{\odot}$')
plt.plot(M40.loc[200:,'log_Teff'],M40.loc[200:,'log_L'],'tab:purple',label='40$M_{\odot}$')
plt.legend(fontsize=14)
plt.xlabel('Log $T_{eff }$', fontsize=18)
plt.ylabel('Log L', fontsize=18)
plt.xlim(4.8,3.9)
```
